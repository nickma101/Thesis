```python
import pandas as pd
from pandas import read_excel
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = read_excel("complete_data_cleaned.xlsx")
df_emotionality = read_excel("complete_data_cleaned_with_emotionality.xlsx")
df_readability = read_excel("complete_data_cleaned_with_readability2.xlsx")
```

```python
df_emotionality["Reach"] = df_emotionality["reach_dummy"]
df_emotionality["Reach"] = df_emotionality["Reach"].replace([0,1], ["regional", "national"])
df_readability["Reach"] = df_readability["reach_dummy"]
df_readability["Reach"] = df_readability["Reach"].replace([0,1], ["regional", "national"])
```

```python
plt.Figure(figsize=(10, 8))
sns.catplot(x="Modality", y="emotionality ratio rauh", palette="pastel", data=df_emotionality, kind="box", showmeans=True, dodge=False)
#sns.swarmplot(x="modality_dummy", y="diversity index", data=df,  color='black', alpha=0.75)
sns.stripplot(x="Modality", y="emotionality ratio rauh", data=df_emotionality,  palette="pastel", jitter = True , alpha=0.05)
plt.ylabel("Emotionality ratio", size=10)
plt.xlabel("Modality", size=10)
plt.title("Emotionality ratio by modality", size=14)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["day"] = df["Date"].dt.day
df["month"] = df["Date"].dt.month
df["year"] = df["Date"].dt.year
```

```python
df["date"] = pd.to_datetime(df[['year', 'month', 'day']])
```

```python
df_online = df[df["Modality"] == "online"]
df_print = df[df["Modality"] == "print"]

df_online = df_online.groupby(df["date"]).count()
df_online = df_online.rename(columns ={"ID":"Number of online articles"})
df_print = df_print.groupby(df["date"]).count()
df_print = df_print.rename(columns ={"ID":"Number of print articles"})
df_print = df_print.iloc[5:]
df_print = df_print.iloc[:-2]
df_print
```

```python
df_online.plot.line(y="Number of articles" , figsize=(12,4))
df_print.plot.line(y="Number of articles" , figsize=(12,4))
```

```python
plt.plot("Number of articles", data=df_online, markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot("Number of articles", data=df_print, color='olive', linewidth=2, linestyle='dashed', label="toto")
```

```python
ax = df_online.plot.line(y="Number of online articles" , figsize=(12,4), title="Number of articles published per day")
df_print.plot(y="Number of print articles" , ax=ax)
```
